```python
#import win32com.client
#from pathlib import Path


#converting doc file to docx
def saveAsDocx(path):
    try:
        #opening ms word
        word = win32.gencache.EnsureDispatch('Word.Application')
    except AttributeError:
        # Remove cache and try again.
        for module in MODULE_LIST:
            if re.match(r'win32com\.gen_py\..+', module):
                del sys.modules[module]
        shutil.rmtree(os.path.join(os.environ.get('LOCALAPPDATA'), 'Temp', 'gen_py'))

    
    #opening ms word
    #word = win32.gencache.EnsureDispatch('Word.Application')
               
    doc = word.Documents.Open(path)
    doc.Activate()
    
    
    #rename path with docx
    new_file_abs = os.path.abspath(path)
    new_file_abs = re.sub(r'\.\w+$','.docx',new_file_abs)
    
    #save & close
    word.ActiveDocument.SaveAs(
        new_file_abs, FileFormat=constants.wdFormatXMLDocument
    )
    doc.Close(False)
    #print('done')
```

```python
def saveAsTextFile(docxFileName):
    #passing docx file to process function
    text = docx2txt.process(docxFileName)

    # Saving content inside docx file into output.txt file
    with open(textFileName, "w", encoding='utf-8') as text_file:
        print(text, file=text_file)
```

```python
def readFromTextFile(textFileName):
    #extract data from txt file
    f = open(textFileName, "r",encoding="utf-8")
    txt_lines = f.readlines() #txt_lines is a list
    #print(txt_lines)#list
    
    #converting list to string
    str_text_data = ' '
    str_text_data = str_text_data.join(txt_lines)
    #print("@@@@@@@@@@@@@")
    #print(str_text_data)
    
    #splitting and converting to list    
    txt_lines = []
    txt_lines = re.split("\t|\n|\s{10,}",str_text_data) #delimiter is tab,newline & 4 spaces
    #print(txt_lines)
    return txt_lines
```

```python
def readFromPdfFile(pdfFileName):
    #extracting from pdf
    pdfdata = fitz.open(pdfFileName)
    pdf_text = ''
    for page in pdfdata:
        pdf_text = pdf_text + str(page.getText())

    #print(pdf_text)
    return pdf_text
```

```python
def readFromDocxFile(docxFileName):
    doc = docx2txt.process(docxFileName)
    #print(doc)
    return doc
```

```python
def process_pdf_data(extracted_data_pdf):
    listData = extracted_data_pdf.split('\n')
    #print(listData)
    #forming a new list by removing empty strings from listData list
    listingData = []

    for i in listData:
        if len(i): #is true when string is empty(contains no whitespace)
            if i.isspace(): #is true when contains only whitespaces or tabs
                continue
            else:
                listingData.append(i)

    #print(len(listingData))
    listingData

    new_list_data = []
    for item in listingData:
        new_list_data.append(item.replace('\t','') and item.strip())

    new_list_data
    
    #removing spaces
    listingData = [x.strip() for x in listingData]
    listingData = [x.replace('\t','') for x in listingData]
    #print(len(listingData))
    return listingData 
    
```

```python
def process_doc_data(extracted_data_doc):
   
    #print(extracted_data_doc)
    list_Data = [x.replace('\n','') for x in extracted_data_doc]
    #print(list_Data)
    listData = [x.replace('\t','') for x in list_Data]
    #print(listData)
    
    #forming a new list by removing empty strings from listData list
    listingData = []

    for i in listData:
        if len(i): #is true when string is empty(contains no whitespace)
            if i.isspace(): #is true when contains only whitespaces or tabs
                continue
            else:
                listingData.append(i)

    #removing spaces
    listingData = [x.strip() for x in listingData]
    
    #print(len(listingData))
    #print(listingData)
    return listingData    
```

```python
restricted_words = ['peoplesoft','admin','administrator','personal','resume','sql','developer','workday','consultant',
                        'page','curriculum','hyderabad','classification', 'internal']
def extract_names(data_list):
    #print(data_list)
    for word in restricted_words:
        for item in data_list:
            if len(item):
                each_item_list = item.lower()
                if word in each_item_list:
                    #print("@@@@",word)
                    #print(each_item_list)
                    #if each_item_list:
                    data_list.remove(item)
                else:
                    continue
            else:
                data_list.remove(item)
    return data_list
```

```python
PHONE_REG = re.compile(r'[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]')
def extract_phone_number(resume_text):
    phone = re.findall(PHONE_REG, resume_text)
    phone_list = []
    if phone:
        for num in phone:
            if len(num) >= 8 and len(num) < 11:
                phone_list.append(num)

    return phone_list
```

```python
#extracting email
EMAIL_REG = re.compile(r'[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+')

def extract_emails(resume_text):
    return re.findall(EMAIL_REG, resume_text)
```

```python
#extracting years of experience
def extract_yearsofexperience():
    experience_list = []
    for i in range(0,len(df)):
        each_row_text = df.loc[i]['Filtered Text'].lower()
        #will check for experience +(year||years||4-digits)
        if re.search(r"\bexperience\b",each_row_text) and (re.search(r"\byear\b",each_row_text) or re.search(r"\byears\b",each_row_text) or re.search(r"\[0-9]{4}",each_row_text)):
            experience_list.append(df.loc[i]['Filtered Text'])
        else:
            pass
                
    return (experience_list)
        
def extract_numbers(list_years):
    exp_num_list = []
    for exp_w in list_years:
        match = re.search(r'(\d+\.?\d*)', exp_w)
        exp_num_list.append(match.group(0))
        
    return exp_num_list
```

```python
def removeMultipleSpaces(input_str):
    return re.sub(r'\s+',' ',input_str) # replaces more than 2 spaces with 1 space
```

```python
# removes text inside brackets including brackets
def removeBrackets(input_str):
    return re.sub(r'\([^()]*\)','',input_str) # removes text inside brackets including brackets
```
